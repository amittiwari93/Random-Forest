```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('Company_Data.csv')
dataset
```

```python
dataset.info()
```

## Converting Sales to category

```python
data=dataset
data.describe()
```

```python
data['Sales']=pd.cut(x=data['Sales'],bins=[0,6,12,17],labels=['Low','Medium','High'],right=False)
data
```

## Splitting data to temp variables

```python
x_temp=data.iloc[:,1:]
y_temp=data.iloc[:,0]
```

```python
x_temp
```

```python
x_temp=pd.get_dummies(x_temp)
```

```python
x_temp
```

## Check the score for individual features

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

bestfeatures=SelectKBest(score_func=chi2,k=5)
fit=bestfeatures.fit(x_temp,y_temp)
# calculate the scores
dfscore=pd.DataFrame(fit.scores_)
# get column names into variable
dfcolumns=pd.DataFrame(x_temp.columns)
# concat 2 dataframe above
feat_col=pd.concat([dfcolumns,dfscore],axis=1)
feat_col.columns=['Specs','Score'] # Naming the DataFrame
print(feat_col.nlargest(5,'Score'))
```

```python
x_temp.head()
```

# Solve with Feature Scaling

   ## Bring every feature on the same scale except ShelveLoc as it is already between 0 and 1

```python
tran_x=x_temp.loc[:,['Price','Advertising','Income','Age','ShelveLoc_Bad','ShelveLoc_Medium','ShelveLoc_Good']]
tran_x
```

```python
y_temp
```

```python
tran_x
```

## Split Train and Test Data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(tran_x, y_temp, test_size = 0.3, random_state = 1)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit_transform(X_train)
sc.fit_transform(X_test)
```

## Try with different estimators and max_depth

```python
from sklearn.ensemble import RandomForestClassifier
# Default values
classifier = RandomForestClassifier(n_estimators = 100, criterion = 'entropy', random_state = 0)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

### Confusion Matrix

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

```python
# estimator 100 max depth 5
classifier = RandomForestClassifier(n_estimators = 100, criterion = 'entropy', random_state = 0,max_depth=5)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

## Gini

```python
classifier = RandomForestClassifier(n_estimators = 100, criterion = 'gini', random_state = 0)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```
